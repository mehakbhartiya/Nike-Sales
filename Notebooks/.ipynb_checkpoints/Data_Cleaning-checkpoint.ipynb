```python
import pandas as pd

url='https://raw.githubusercontent.com/mehakbhartiya/Nike-Sales/refs/heads/main/Data/Nike_Sales_Cleaned.csv'
df = pd.read_csv(url)

df.head()
df.info()
```

```python
print(df['Order_Date'].head(10))
print(df['Order_Date'].dtype)
```

Since the Order_Date type are different we need to clean it

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
```

#### After Cleaning

```python
print(df['Order_Date'].head(10))
print(df['Order_Date'].dtype)
```

```python
df['Order_Date'].isna().sum()
```

Here the Order_date is converted to datetime and if the order_date cant be converted it is converted to NAT

```python
df[df['Units_Sold'] < 0]
```

In the above Query it shows that unit sold can be negative which will give error while procedding for analysis

```python
(df['Units_Sold'] < 0).sum()
```

Here 205 rows contain the negative values which leads to wrong analysis so we need to fix this.

```python
df = df[df['Units_Sold'] >= 0]
```

#### After Cleaning

```python
(df['Units_Sold'] < 0).sum()
```

Now let’s check if there are any dicounts more than 1. If so, it means it’s referring to a discount above 100%, meaning that the store would pay the customer to take the product.

```python
df[df['Discount_Applied'] > 1]
```

```python
(df['Discount_Applied'] > 1).sum()
```

73 rows can similar issue

```python
df = df[df['Discount_Applied'] <= 1]
```

#### After Cleaning

```python
(df['Discount_Applied'] > 1).sum()
```

Dropping the duplicate ids

```python
df = df.drop_duplicates(subset='Order_ID')
```

```python
df['Region'].value_counts().head(30)
```

How many distinct Regions are there in dataset

```python
sorted(df['Region'].dropna().astype(str).str.strip().unique())
```

```python
df['Region'] = (
    df['Region'].astype(str).str.strip().str.lower()
      .replace({'hyd': 'hyderabad', 'hyderbad': 'hyderabad', 'bengaluru': 'bangalore'})
      .str.title()
)
```

#### After Cleaning

```python
sorted(df['Region'].dropna().astype(str).str.strip().unique())
```

```python
df[['MRP','Units_Sold']].isnull().sum()
```

```python
mrp_median = df['MRP'].median()
df['MRP'] = df['MRP'].fillna(mrp_median)
```

```python
df[['MRP','Units_Sold']].isna().sum()
```

```python
df.to_csv("Nike_Sales_Cleaned.csv", index=False)
```
